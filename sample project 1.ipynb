## OUTLIER REMOVAL 

## 1. SALES

```python
df["log_base2_sales"]=np.log2(df["Sales"])
print("After logarithmic transformation the modified dataset is: ")
df
```

```python
print( "Box plot of log_base2_sales:\n \n")
plt.figure(figsize=(10,8))
sns.boxplot(data=df, x="log_base2_sales")
plt.title("log_base2_sales", size=25)
plt.show()
```

### After logarithmic transformation we can see that few points are outside of boxplot. 

## 2. PROFIT

```python
df["log_base2_profit"]=np.log2(df["Profit"])
print("After logarithmic transformation the modified dataset is: ")
df
```

```python
print( "Box plot of log_base2_profit:\n \n")
plt.figure(figsize=(10,8))
sns.boxplot(data=df, x="log_base2_profit")
plt.title("log_base2_profit", size=25)
plt.show()
```

## 3. QUANTITY

```python
df["log_base2_quantity"]=np.log2(df["Quantity"])
print("After logarithmic transformation the modified dataset is: ")
df
```

```python
print( "Box plot of log_base2_quantity:\n \n")
plt.figure(figsize=(10,8))
sns.boxplot(data=df, x="log_base2_quantity")
plt.title("log_base2_quantity", size=25)
plt.show()
```

### 12. Region vs Profit (Pie chart)

```python
plt.figure(figsize=(12,10))

R= df.groupby(by="Region")["Profit"].sum()

R.plot( kind="pie",
          colors =["lime","magenta","yellow","blue"],
          shadow= True,
          startangle=180,
          explode=(0.2,0.0,0.0,0.08),
          autopct ="%1.1f%%"
        )
                                            

plt.title("Pie diagram representing profit based on regions",
           size=25,
           color="Maroon"
         )

plt.show()
```

#### Observation : From the above pie chart we can see that central region has lowest profit and west region has the highest profit. So if we want to increase the profit we have to focus on central region.
